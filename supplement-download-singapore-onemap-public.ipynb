```python
def get_openmapsg_response(search_val, token=token):
    url = f"https://www.onemap.gov.sg/api/common/elastic/search?searchVal={search_val}&returnGeom=Y&getAddrDetails=Y"
    headers = {"Authorization": token}
    response = requests.request("GET", url, headers=headers)
    return response

def get_openmapsg_result(response):
    return response.json()['results']

def get_multiple_results(vals: list, data_path, json_results_file, token=token,):
    results = []
    exceptions = []
    time_started = datetime.now()
    last_retrieve_time = datetime.now()
    for v in vals:
        print(f"Rerieving: ", v)
        try:
            response = get_openmapsg_response(v, token)
            result = get_openmapsg_result(response)
            results.append({'name': v, 'data': result})
            time_done = datetime.now()
            time_used = (time_done - last_retrieve_time).seconds
            total_time_used = (time_done - time_started).seconds
            last_retrieve_time = time_done
            with open(data_path+json_results_file, "w") as final:
                json.dump(results, final)
            print(f"{v} done in {time_used} sec. Total time used: {total_time_used} sec ")
        except:
            print('Error: ', v)
            exceptions.append(v)
    return (results, exceptions)
        
```

```python

def populate_project_list(positive_results):
    projects_list_v1 = []
    count = 0
    for res in positive_results:
        name = res['name']
        project_entries_list = res['data']
        if len(project_entries_list) == 0:
            print("Missing value: ", count)
            continue
        x_list = []
        y_list = []
        lon_list = []
        lat_list = []

        for entry in project_entries_list:
            x_list.append(entry['X'])
            y_list.append(entry['Y'])
            lon_list.append(entry['LONGITUDE'])
            lat_list.append(entry['LATITUDE'])

        projects_list_v1.append((name, 
                              np.array(x_list).astype(np.float).mean(), 
                              np.array(y_list).astype(np.float).mean(), 
                              np.array(lat_list).astype(np.float).mean(), 
                              np.array(lon_list).astype(np.float).mean()))
        print(f'{count}', end="\r")
        count += 1
    return projects_list_v1
```
